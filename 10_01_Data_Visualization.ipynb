# Data visualization

An essential part of the scientific process is visualizing the results you have produced.

For example, when setting out to answer a research question, a methodology is usually specified. After this methodology has been utilized, the results need to be reported.

You may generate lots and lots of results, as this is part of the research process. After which, you may have to select the best data to present as part of the scientific narrative of your paper. These data items require plotting.

The easiest way to plot your analysis results is via `Pandas`, as covered here.


# Plotting in Pandas

Once you have your data in a `Pandas` dataframe, you only need a couple of extra lines to begin generating high-quality visualizations.

Plotting capabilities in Pandas are built on top of Matplotlib, a widely used Python plotting library.

Pandas provides the method `.plot()` which can be used on both Series and DataFrame objects. This method creates a variety of standard visualizations by interpreting the underlying data.

Common types of plots include the following:

```python
# Example: Line Plot (default) for visualizing trends over time
# or relationships between data points
import pandas as pd
import matplotlib.pyplot as plt

# Sample Data: Population density (people per square km) over years
data = {
    'Year': [2010, 2012, 2014, 2016, 2018, 2020],
    'Region_A': [120, 125, 130, 135, 140, 145],
    'Region_B': [80, 85, 88, 90, 95, 100],
    'Region_C': [200, 205, 210, 220, 230, 240]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Set the 'Year' column as the index for better plotting
df.set_index('Year', inplace=True)

# Plot the data
df.plot(kind='line', title='Population Density Over Time', figsize=(8, 6))

# Customize the plot
plt.xlabel('Year')
plt.ylabel('Population Density (people/kmÂ²)')
plt.grid(True)
plt.legend(title='Regions')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Example: Bar Chart: Useful for comparing categorical data
import pandas as pd
import matplotlib.pyplot as plt

# Sample Data: Land cover (in square kilometers) for different regions
data = {
    'Region': ['Region_A', 'Region_B', 'Region_C', 'Region_D'],
    'Forest': [150, 120, 100, 80],
    'Agriculture': [200, 180, 150, 120],
    'Urban': [50, 70, 100, 150]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Set the 'Region' column as the index for better plotting
df.set_index('Region', inplace=True)

# Plot the data
df.plot(kind='bar', title='Land Cover Distribution by Region', figsize=(8, 6))

# Customize the plot
plt.xlabel('Region')
plt.ylabel('Land Cover (kmÂ²)')
plt.xticks(rotation=0)  # Rotate x-axis labels for readability
plt.legend(title='Land Cover Type')
plt.grid(axis='y')  # Add horizontal grid lines for better comparison
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Example: Histogram: Displays the distribution of a numerical variable
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Sample Data: Elevation values (in meters) from a region
elevation_data = { # Simulated elevation data
    'Elevation': np.random.normal(loc=500, scale=100, size=1000)
}

# Create a DataFrame
df = pd.DataFrame(elevation_data)

# Plot the histogram
df['Elevation'].plot(
    kind='hist',
    bins=20,
    title='Elevation Distribution in Region',
    figsize=(8, 6),
    alpha=0.7
    )

# Customize the plot
plt.xlabel('Elevation (meters)')
plt.ylabel('Frequency')
# Add horizontal grid lines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Example: Box Plot: Summarizes the distribution and variability of data
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Sample Data: Annual rainfall (in mm) for different regions
np.random.seed(42)  # Set seed for reproducibility
data = { # Simulated rainfall data
    'Region_A': np.random.normal(loc=800, scale=100, size=50),
    'Region_B': np.random.normal(loc=600, scale=120, size=50),
    'Region_C': np.random.normal(loc=1000, scale=150, size=50),
    'Region_D': np.random.normal(loc=750, scale=90, size=50)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Plot the box plot
df.plot(
    kind='box',
    title='Annual Rainfall Distribution Across Regions',
    figsize=(8, 6)
    )

# Customize the plot
plt.ylabel('Rainfall (mm)')
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add horizontal grid lines
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Example: Scatter Plot: Shows relationships between two variables
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Sample Data: Elevation and Temperature for various locations
# Set seed for reproducibility
np.random.seed(42)
# Elevation in meters
elevation = np.random.uniform(100, 2000, 100)
# Temperature (Â°C) decreases with elevation
temperature = 30 - 0.005 * elevation + np.random.normal(0, 2, 100)

# Create a DataFrame
data = {
    'Elevation': elevation,
    'Temperature': temperature
}
df = pd.DataFrame(data)

# Plot the scatter plot
df.plot(
    kind='scatter',
    x='Elevation',
    y='Temperature',
    title='Temperature vs. Elevation',
    figsize=(8, 6),
    alpha=0.7
    )

# Customize the plot
plt.xlabel('Elevation (meters)')
plt.ylabel('Temperature (Â°C)')
plt.grid(True, linestyle='--', alpha=0.7)  # Add a grid for clarity
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Example: Stacked Bar Chart: Useful for comparing proportions or
# distributions across categories
import pandas as pd
import matplotlib.pyplot as plt

# Sample Data: Land cover proportions (in percentage) for different regions
data = {
    'Region': ['Region_A', 'Region_B', 'Region_C', 'Region_D'],
    'Forest': [40, 35, 25, 20],
    'Agriculture': [30, 40, 50, 30],
    'Urban': [30, 25, 25, 50]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Set the 'Region' column as the index for better plotting
df.set_index('Region', inplace=True)

# Plot the stacked bar chart
df.plot(kind='bar', stacked=True, figsize=(8, 6), title='Land Cover Proportions by Region')

# Customize the plot
plt.xlabel('Region')
plt.ylabel('Land Cover Proportion (%)')
plt.xticks(rotation=0)  # Keep x-axis labels horizontal
plt.legend(title='Land Cover Type')
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add horizontal grid lines for better readability
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Example: Develop a panel plot with Pandas
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Data preparation for each plot
# 1. Line Plot Data
line_data = {
    'Year': [2010, 2012, 2014, 2016, 2018, 2020],
    'Region_A': [120, 125, 130, 135, 140, 145],
    'Region_B': [80, 85, 88, 90, 95, 100],
    'Region_C': [200, 205, 210, 220, 230, 240]
}
line_df = pd.DataFrame(line_data).set_index('Year')

# 2. Bar Chart Data
bar_data = {
    'Region': ['Region_A', 'Region_B', 'Region_C', 'Region_D'],
    'Forest': [150, 120, 100, 80],
    'Agriculture': [200, 180, 150, 120],
    'Urban': [50, 70, 100, 150]
}
bar_df = pd.DataFrame(bar_data).set_index('Region')

# 3. Histogram Data
elevation_data = np.random.normal(loc=500, scale=100, size=1000)

# 4. Box Plot Data
box_data = {
    'Region_A': np.random.normal(loc=800, scale=100, size=50),
    'Region_B': np.random.normal(loc=600, scale=120, size=50),
    'Region_C': np.random.normal(loc=1000, scale=150, size=50),
    'Region_D': np.random.normal(loc=750, scale=90, size=50)
}
box_df = pd.DataFrame(box_data)

# 5. Scatter Plot Data
elevation = np.random.uniform(100, 2000, 100)
temperature = 30 - 0.005 * elevation + np.random.normal(0, 2, 100)
scatter_df = pd.DataFrame({'Elevation': elevation, 'Temperature': temperature})

# 6. Stacked Bar Plot Data
stacked_data = {
    'Region': ['Region_A', 'Region_B', 'Region_C', 'Region_D'],
    'Forest': [40, 35, 25, 20],
    'Agriculture': [30, 40, 50, 30],
    'Urban': [30, 25, 25, 50]
}
stacked_df = pd.DataFrame(stacked_data).set_index('Region')

# Create the panel plot
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(8, 11))
axes = axes.flatten()  # Flatten the array for easy indexing

# Line Plot
line_df.plot(ax=axes[0], title='(A) Line Plot: Population Density')
axes[0].set_ylabel('Density (people/kmÂ²)')
axes[0].set_xlabel('Year')

# Bar Chart
bar_df.plot(kind='bar', ax=axes[1], title='(B) Bar Chart: Land Cover')
axes[1].set_ylabel('Area (kmÂ²)')
axes[1].set_xlabel('Region')
axes[1].set_xticklabels(bar_df.index, rotation=0)

# Histogram
axes[2].hist(elevation_data, bins=20, alpha=0.7)
axes[2].set_title('(C) Histogram: Elevation Distribution')
axes[2].set_xlabel('Elevation (meters)')
axes[2].set_ylabel('Frequency')

# Box Plot
box_df.plot(kind='box', ax=axes[3], title='(D) Box Plot: Rainfall Distribution')
axes[3].set_ylabel('Rainfall (mm)')

# Scatter Plot
axes[4].scatter(scatter_df['Elevation'], scatter_df['Temperature'], alpha=0.7)
axes[4].set_title('(E) Scatter Plot: Temp vs Elevation')
axes[4].set_xlabel('Elevation (meters)')
axes[4].set_ylabel('Temperature (Â°C)')

# Stacked Bar Plot
stacked_df.plot(kind='bar', stacked=True, ax=axes[5], title='(F) Stacked Bar Plot: Land Cover Proportions')
axes[5].set_ylabel('Proportion (%)')
axes[5].set_xlabel('Region')
axes[5].set_xticklabels(stacked_df.index, rotation=0)

# Adjust layout for better readability
plt.tight_layout()

# Save the panel plot as a PNG file
plt.savefig('panel_plot_spatial_computing.png', dpi=300, bbox_inches='tight')
plt.show()
```

# Plotting in GeoPandas

It is equally easy to plot in GeoPandas, as demonstrated here.

Firstly, we will cover plotting of points, and then secondly, we will cover lines.

```python
# Example: Plotting points with GeoPandas
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point
import numpy as np

# Generate example points with random coordinates and values
np.random.seed(42)
points_data = {
    "Location": [f"Point {i+1}" for i in range(15)],
    "Value": np.random.randint(50, 500, size=15),  # Random values
    "geometry": [Point(x, y) for x, y in zip(
        np.random.uniform(-120, -60, 15),  # Random longitudes
        np.random.uniform(20, 50, 15)     # Random latitudes
    )]
}

# Convert the data into a GeoDataFrame
points_gdf = gpd.GeoDataFrame(points_data, crs="EPSG:4326")

# Plot the points
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
points_gdf.plot(column="Value",  # Column to color by
                cmap="coolwarm",  # Color map
                legend=True,  # Show legend
                legend_kwds={"label": "Point Value", "shrink": 0.6},  # Legend label
                markersize=points_gdf["Value"],  # Scale marker size by value
                ax=ax)

# Add title and axis labels
ax.set_title("Points with Random Values", fontsize=16)
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")

# Save and show the plot
plt.savefig("points_plot.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Example: Plotting linestrings with GeoPandas
from shapely.geometry import LineString

# Generate example lines with random start and end points
np.random.seed(42)
lines_data = {
    "Route": [f"Route {i+1}" for i in range(15)],
    "Distance": np.random.randint(100, 1000, size=15),  # Random distances
    "geometry": [
        LineString([(x1, y1), (x2, y2)]) for x1, y1, x2, y2 in zip(
            np.random.uniform(-120, -60, 15),  # Random longitudes (start)
            np.random.uniform(20, 50, 15),    # Random latitudes (start)
            np.random.uniform(-120, -60, 15),  # Random longitudes (end)
            np.random.uniform(20, 50, 15)     # Random latitudes (end)
        )
    ]
}

# Convert the data into a GeoDataFrame
lines_gdf = gpd.GeoDataFrame(lines_data, crs="EPSG:4326")

# Plot the lines
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
lines_gdf.plot(column="Distance",  # Column to color by
               cmap="viridis",  # Color map
               legend=True,  # Show legend
               legend_kwds={"label": "Line Distance (km)", "shrink": 0.6},
               linewidth=2,  # Line thickness
               ax=ax)

# Add title and axis labels
ax.set_title("15 Lines with Random Distances", fontsize=16)
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")

# Save and show the plot
plt.savefig("lines_plot.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Example: Developing a panel plot with GeoPandas
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point, LineString
import numpy as np

# Generate 15 example points with random coordinates and values
np.random.seed(42)
points_data = {
    "Location": [f"Point {i+1}" for i in range(15)],
    "Value": np.random.randint(50, 500, size=15),  # Random values
    "geometry": [Point(x, y) for x, y in zip(
        np.random.uniform(-120, -60, 15),  # Random longitudes
        np.random.uniform(20, 50, 15)     # Random latitudes
    )]
}
points_gdf = gpd.GeoDataFrame(points_data, crs="EPSG:4326")

# Generate 15 example lines with random start and end points
lines_data = {
    "Route": [f"Route {i+1}" for i in range(15)],
    "Distance": np.random.randint(100, 1000, size=15),  # Random distances
    "geometry": [
        LineString([(x1, y1), (x2, y2)]) for x1, y1, x2, y2 in zip(
            np.random.uniform(-120, -60, 15),  # Random longitudes (start)
            np.random.uniform(20, 50, 15),    # Random latitudes (start)
            np.random.uniform(-120, -60, 15),  # Random longitudes (end)
            np.random.uniform(20, 50, 15)     # Random latitudes (end)
        )
    ]
}
lines_gdf = gpd.GeoDataFrame(lines_data, crs="EPSG:4326")

# Create a panel plot with one column and two rows
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 11))

# Plot Points (A)
points_gdf.plot(column="Value",  # Column to color by
                cmap="coolwarm",  # Color map
                legend=True,  # Show legend
                legend_kwds={"label": "Point Value", "shrink": 0.6},
                markersize=points_gdf["Value"],  # Scale marker size by value
                ax=axes[0])
axes[0].set_title("(A) Points with Random Values", fontsize=16)
axes[0].set_xlabel("Longitude")
axes[0].set_ylabel("Latitude")

# Plot Lines (B)
lines_gdf.plot(column="Distance",  # Column to color by
               cmap="viridis",  # Color map
               legend=True,  # Show legend
               legend_kwds={"label": "Line Distance (km)", "shrink": 0.6},
               linewidth=2,  # Line thickness
               ax=axes[1])
axes[1].set_title("(B) Lines with Random Distances", fontsize=16)
axes[1].set_xlabel("Longitude")
axes[1].set_ylabel("Latitude")

# Adjust layout
plt.tight_layout()

# Save the panel plot as a PNG
plt.savefig("panel_plot_points_lines.png", dpi=300, bbox_inches="tight")
plt.show()
```
