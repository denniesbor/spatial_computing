# Plotting in Pandas

Once you have your data in a `Pandas` dataframe, you can very easily plot your data.

Below are a set of plotting examples which you could utilize for your assignment project.

First, each example specifies the input data (which would need adapting for each context), and converts this to a `Pandas` df for plotting.

Secondly, we specify the call to the `Pandas` plotting function.

Let us begin with a basic bar plot.

```python
# Example: First, we create a basic df for plotting
import pandas as pd
import matplotlib.pyplot as plt

# We can create the underlying data for our Pandas df
# Here we have route scenarios, and the estimated travel time
data_raw = {
    'route': ['A', 'B', 'C', 'D'],
    'time_hours': [1.5, 0.5, 1, 0.75]
  }

# Convert to a pandas df
data = pd.DataFrame(data_raw)
data
```

Now we can call our plotting function using `.plot.bar()`.

The in-line code comments should be self explanatory for each argument, e.g., the x axis column, or y axis column, title, etc.

```python
# Now we can plot our barplot data
data.plot.bar(
    x='route',                          # x axis variable
    y='time_hours',                     # y axis variable
    title="Barplot of Data Values",     # plot title
    rot=0                               # rotate x axis labels
  )
plt.xlabel('Route Scenario')
plt.ylabel('Travel Time (hours)')
```

Now you have the hang of the approach we will take (specifying our plotting data, and then plotting), it should be easier to understand how to create more complicated plots.

Next, we will look at a dodged barplot containing travel times for either walking or driving via car.  

```python
# Example: Creating a dodged barplot
import pandas as pd
import matplotlib.pyplot as plt

# Define our variables
walking = [1.5, 2.5, 1.7, 0.75]
car = [0.3,0.5,0.4,0.2]
route = ['A', 'B', 'C', 'D']

# Create our df
df = pd.DataFrame({'walking': walking,
                   'car': car}, index=route)

# Finally, we can plot the data
df.plot.bar(
    title="Dodged Barplot of Travel Time Data Values",
    rot=0
)
plt.xlabel('Route Scenario')
plt.ylabel('Travel Time (hours)')
```

If we want, we can also create a panel plot, placing each of these data columns on their own separate subplot graph.

```python
# Example: Creating a panel plot
import pandas as pd
import matplotlib.pyplot as plt

# Define our variables
walking = [1.5, 2.5, 1.7, 0.75]
car = [0.3,0.5,0.4,0.2]
route = ['A', 'B', 'C', 'D']

# Create our df
df = pd.DataFrame({'walking': walking,
                   'car': car}, index=route)

# Finally, we can plot the data
df.plot.bar(
    title="Panel plot of Travel Time Data Values",
    rot=0,
    subplots=True
)
plt.xlabel('Route Scenario')
plt.ylabel('Travel Time (hours)')
```

Finally, we can consider a stacked barplot.

Stacked barplots are very useful for showing the composition of a set of results.

For example, here we plot the potential damage cost by category for a volcano eruption radius.

```python
# Example: Creating a stacked barplot
import pandas as pd
import matplotlib.pyplot as plt

# Define our variables
buildings_mn = [5,8,24,36,43]
roads_mn = [1,2,3,4,6]
distance = ['5 km','10 km','15 km','20 km','25 km']

# Create our df
df = pd.DataFrame({'buildings': buildings_mn,
                   'roads': roads_mn}, index=distance)

# Finally, we can plot the data
df.plot.bar(
    title="Volcano Damage Cost",
    rot=0,
    stacked=True
)
plt.xlabel('Volcano Eruption Damage Radius (km)')
plt.ylabel('Replacement Cost (US$ Mn)')
```
